```python
# Projektaufgabe von Erdenetuya und Xuantong
import numpy as np
import matplotlib.pyplot as plt
from interpol1d import splineinterpol, poly_interpol
```

```python
# Aufgabe 2
def f(x):
    return 1 /(1 + x*x)

n = 100
f_x = np.linspace(-5, 5, n)
f_y = f(f_x)
   
number_of_support_points=5
xi = np.linspace(-5, 5, number_of_support_points)
yi = f(xi)
plt.plot(xi, yi, '*', markersize=10) # xi, yi

plt.plot(f_x, f_y, linewidth = 2) # f(x)

for i in range(1, number_of_support_points):
    x = np.linspace(xi[i-1], xi[i], n)
    plt.plot(x, splineinterpol(x, xi ,yi), linewidth=2, color='red') # SplinInterpol
    plt.plot(x, poly_interpol(x, xi, yi), linewidth=2, color='green') # PolyInterpol
    
plt.legend(['', r'$f(x)$', 'SplinInterpol', 'PolyInterpol'])
plt.xlabel('x')
plt.ylabel('y')
```

```python
# doppeltlogarithmischer Darstellung des experimentellen Fehlers über ℎ 

errors = []
hs = []
for number_of_support_points in range(3, 55, 2): # Je größer number_of_support_points ist, destor kleiner ℎ wird.

    xi = np.linspace(-5, 5, number_of_support_points)
    yi = f(xi)

    hs.append(xi[1]-xi[0])
    error = 0
    for i in range(1, number_of_support_points):
        x = np.linspace(xi[i-1], xi[i], n)
        s = splineinterpol(x, xi ,yi) # SplinInterpol
        y = f(x)
        error = max(error, np.linalg.norm(y-s, ord=np.inf))
    errors.append(error)
plt.plot(hs, errors, linewidth=2)
plt.xscale("log")
plt.gca().invert_xaxis()
plt.yscale("log")
plt.xlabel('h')
plt.ylabel('error')
```
