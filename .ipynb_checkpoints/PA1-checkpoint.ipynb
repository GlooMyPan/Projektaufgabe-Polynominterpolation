# Aufgabe 2
Vergleichung die Approximationseigenschaften der kubischen Spline-Interpolation und der Polynominterpolation für die Rungefunktion.

```python
# Projektaufgabe von Erdenetuya Undral und Xuantong Pan
import numpy as np
import matplotlib.pyplot as plt
from interpol1d  import splineinterpol, poly_interpol
```

```python
# Aufgabe 2
def f(x):
    return 1 /(1 + x*x)

n = 100
f_x = np.linspace(-5, 5, n)
f_y = f(f_x)
   
number_of_support_points=5
xi = np.linspace(-5, 5, number_of_support_points)
yi = f(xi)
plt.plot(xi, yi, '*', markersize=10) # xi, yi

plt.plot(f_x, f_y, linewidth = 2) # f(x)

for i in range(1, number_of_support_points):
    x = np.linspace(xi[i-1], xi[i], n)
    plt.plot(x, splineinterpol(x, xi ,yi), linewidth=2, color='red') # SplinInterpol
    plt.plot(x, poly_interpol(x, xi, yi), linewidth=2, color='green') # PolyInterpol
    
plt.legend(['', r'$f(x)$', 'SplinInterpol', 'PolyInterpol'])
plt.xlabel('x')
plt.ylabel('y')
```

```python
# doppeltlogarithmischer Darstellung des experimentellen Fehlers über ℎ 

errors = []
hs = []
for number_of_support_points in range(3, 55, 2): # Je größer number_of_support_points ist, destor kleiner ℎ wird.

    xi = np.linspace(-5, 5, number_of_support_points)
    yi = f(xi)

    hs.append(xi[1]-xi[0])
    error = 0
    for i in range(1, number_of_support_points):
        x = np.linspace(xi[i-1], xi[i], n)
        s = splineinterpol(x, xi ,yi) # SplinInterpol
        y = f(x)
        error = max(error, np.linalg.norm(y-s, ord=np.inf))
    errors.append(error)
plt.plot(hs, errors, linewidth=2)
plt.xscale("log")
plt.gca().invert_xaxis()
plt.yscale("log")
plt.xlabel('h')
plt.ylabel('error')
```

## Polynominterpolation

**Ziel:** Wir suchen einem Polynom $n$-ten Grades $p(x)$ aus, wechel durch vorgegebene Stützstellen $x_i$ und Stützwerten $y_i$ verläuft. 

#### Idee: Interpolationspolynom mit Newtonbasis sukzessive aufzubauen.
**1.Schritt:** Suche Basis ${\omega_0, ..., \omega_n}$ des Basis $\prod_n$ mit $\omega_0 = 1, \omega_i(x) = \prod_{j=0}^{i-1}(x-x_j), i = 1, ..., n$ und Koeffizienten $b_0, ..., b_n$. Mit den  können wir eine Darstellung des Interpolationspolynom in dieser Basis bilden, und diese Darstellung heißt Newtondarstellung.

**2.Schritt:** Schreiben wir $b_n$ mit Parameter $y_n, P(f|x_0, ..., x_n)(x_n)  und \omega_n(x_n)$ um, also $b_i =: f[x_0, ..., x_i]$.

**3.Schritt:** Wir schreiben $f[x_r, ..., x_s]$ in einen Rekursionsformel.

**4.Schritt:** Rechnen wir $b_0,..., b_n$ mit diese Rekursionsformel aus dann in den Newtondarstellung einsezten, bekommen wir dann den Interpolationspolynom.

## Spline-Interpolation

**Ziel:** Zwischen je 2 Stützstellen rechnet man ein kubisch Polynom aus, dass in jede $s(x_i)$ genau $y_i$ rauskommt. Und alle Kubischen Polynomen bilden eine Kurve von $x_0$ bis $x_n$ durch alle Stützwerten $y_i$

**1. Schritt:** Angenomen $m_j$ die zweite Ableitung von $s_j$($x_i$) ist, wissen wir schon, dass $m_0$ und $m_n$ gleich 0 sind(natürliche Randbedingung). Die werte $m_1$, ..., $m_{n-1}$ sind unbekannt.

**2. Schritt:** Wir bilden $s_j^{\prime\prime}(x)$ mit $m_j$ in Lagrangedarstellung, nach zweifache Intergration bekommen wir die Gleichung von $s_j(x)$

**3. Schritt:** Wegen $s_{j-1}^{\prime}(x_j) = s_j^{\prime}(x_j), j = 1, ..., n-1$, können wir eine Gleichung von $m_{j-1}, m_{j}, m_{j+1}, j = 1, ..., n-1$ bilden, mit der Gleichung können wir $m_j, j = 1, ..., n-1$ durch LGS ausrechnen. Dann bekommen wir diesen kubischen Polynomen $s_0(x), ..., s_n(x)$

### Implementieren
##### Polynominterpolation
1. Berechnen wir für die gegebenen Stützpunkten die Koeffizienten $b_0, .. b_n$ mit dem Rekursionsformel.
2. Das Interpolationspolynom wird mit den berechneten Koeffizienten und den gegebenen Stützstellen durch dem Hornerschema ausgewertet.

##### Splineinterpolation
1. Erstellen der Diagonalmatrix A.
2. Aufbauen des LGS (Ableitungen).
3. Lösen des LGS 
4. Auswertung des kubischen Polynoms durch splinej().
